```python
import pandas_datareader.data as web
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import date, timedelta
%matplotlib inline
```

```python
start_date = (date.today() - timedelta(days=3650)).strftime('%Y-%m-%d')
end_date = date.today().strftime('%Y-%m-%d')
year_ago = (today - timedelta(days=365)).strftime('%Y-%m-%d')
```

```python
# Bank of America
BAC = yf.download('BAC', start=start_date, end=end_date, progress=False)

# CitiGroup
C = yf.download('C', start=start_date, end=end_date, progress=False)

# Goldman Sachs
GS = yf.download('GS', start=start_date, end=end_date, progress=False)

# JPMorgan Chase
JPM = yf.download('JPM', start=start_date, end=end_date, progress=False)

# Morgan Stanley
MS = yf.download('MS', start=start_date, end=end_date, progress=False)

# Wells Fargo
WFC = yf.download('WFC', start=start_date, end=end_date, progress=False)

#RBC
RY = yf.download('RY', start=start_date, end=end_date, progress=False)
```

```python
BAC.head()
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC', 'RY']
```

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC, RY],axis=1,keys=tickers)
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.drop('2013-09-09	', inplace=True)
```
